```python
from kbhit import KBHit
class Osc:
    def __init__(self, freq = 440, volume = 1, phase = 0):
        self.freq = freq
        self.volume = volume
        self.phase = phase
        self.frame = -0


    def next(self):
        bloque = np.float32(np.arange(self.frame, self.frame+CHUNK))
        self.frame+=CHUNK
        bloque = np.sin(self.phase + bloque*2*np.pi*self.freq/SRATE)
        bloque*=self.volume
        return bloque

    def getVolume(self):
        return self.volume
    def setVolume(self, volume):
        self.volume = volume
        if self.volume < 0:
            self.volume = 0
        elif self.volume > 1:
            self.volume = 1
        
    def getFrequency(self):
        return self.freq
    def setFrequency(self, freq):
        self.freq = freq

class Modulator(Osc):
    def __init__(self, freq=440, volume=1, phase=0, min=0, max=1):
        Osc.__init__(self, freq, volume, phase)
        self.min = min
        self.max = max
    
    def next(self):
        medio=(self.min+self.max)/2
        modulator=Osc.next(self)*medio+medio
        return modulator

class Paneo(Modulator):
    def __init__ (self, freq=440, volume=1, phase=0, min=0, max=1):
        Modulator.__init__(self, freq, volume, phase, min, max)
        self.left=1
        self.right=1

    def pan(self, direction):
        self.left-=0.1*direction
        if self.left <= self.min:
            self.left=self.min
        elif self.left>=self.max:
            self.left=self.max
        self.right+=0.1*direction
        if self.right<= self.min:
            self.right=self.min
        elif self.right>=self.max:
            self.right=self.max
    
    def next(self):
        l = self.left*Modulator.next(self)
        r = self.right*Modulator.next(self)
        return np.column_stack((l,r))
    
def testPaneo():
    mod = Paneo()
    input = KBHit()
    end = False
    stream = sd.OutputStream(samplerate=SRATE, blocksize=CHUNK, channels=2)
    stream.start()
    while not(end):
        bloque = mod.next()
        key = input.getKey()
        print(key)
        if key != '':
            end = (key=='escape')
            if key=='p':
                mod.pan(-1)
            elif key=='o':
                mod.pan(1)
        stream.write(bloque)
        # print("Write")
        
    input.quit()
    stream.stop()
    stream.close()

testPaneo()



        





    
```
